```python
from confluent_kafka import Producer
import json
import time
import random
from faker import Faker
```

```python
fake = Faker()
conf = {
    'bootstrap.servers': 'Replace it with your bootstrap server.',
    'security.protocol': 'SASL_SSL',
    'sasl.mechanism': 'PLAIN',
    'sasl.username': 'Replace it with your API key.',
    'sasl.password': 'Replace it with your API secret'
}
```

```python
producer = Producer(conf)
def generate_transactions():
    transaction = {
        "transaction_id" : fake.uuid4(),
        "timestamp":int(time.time()),
        "user_id":random.randint(10000,99999),
        "amount":round(random.uniform(5,5000),2),
        "transaction_type":random.choice(["purchase","transfer","withdrawal"]),
        "location":fake.city(),
        "mearchant":fake.company(),
        "card_number":fake.credit_card_number()
    }
    return transaction
def delivery_report(err,msg):
    if err is not None:
        print(f"message is not delivered: {err}")
    else:
        print(f"msg delivered to {msg.topic()},{[{msg.partition()}]},{msg}")

while True :
    txn = generate_transactions()
    producer.produce("txndata",value = json.dumps(txn),key = str(txn['user_id']),callback = delivery_report)
    producer.flush()
    time.sleep(1)
```
